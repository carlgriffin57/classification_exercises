```python
import pandas as pd
```

1. Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
att_df = pd.read_csv('untidy-data/attendance.csv')
```

```python
att_df
```

```python
att_df.info()
```

```python
att_df.replace(to_replace={'P': 1.00,
                                  'T': 0.90,
                                  'H': 0.50,
                                  'A': 0.00},
                      inplace=True)
```

```python
att_df
```

```python
att_df = att_df.rename(columns={'Unnamed: 0':'student'})
```

```python
att_df
```

```python
att_df = att_df.melt(id_vars=['student'], var_name='date', value_name='attendance_score')
```

```python
att_df.info()
```

```python
att_df.groupby('student').attendance_score.mean()
```

3. Cake Recipes

Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.


Tidy the data as necessary.

Which recipe, on average, is the best? recipe b


Which oven temperature, on average, produces the best results? 275

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_df = pd.read_csv('untidy-data/cake_recipes.csv')
```

```python
cake_df
```

```python
cake_df.info()
```

```python
 cake_df = cake_df.rename(columns={'recipe:position':'recipe'})
```

```python
cake_df
```

```python
cake_df[['recipe', 'position']] = cake_df.recipe.str.split(':', expand=True)
```

```python
cake_df
```

```python
cake_long = cake_df.melt(id_vars=['recipe', 'position'], var_name='temperature', value_name='tastiness')
```

```python
cake_long.head()
```

Which recipe, on average, is the best? recipe b


```python
cake_long.groupby('recipe').tastiness.mean()
```

Which oven temperature, on average, produces the best results? 275

```python
cake_long.groupby('temperature').tastiness.mean()
```

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_long.sort_values(by='tastiness', ascending=False)
```
